```python
import pandas as pd
import re
```

```python
df_players = pd.read_json('./Data/liga brasil/players.json')
df_players.head()
```

```python
df_players['name'] = df_players['code'].str.replace('-', ' ')
df_players = df_players.drop('last_name',axis=1)
df_players = df_players.drop('code',axis=1)
```

```python
df_players.info()
```

```python
# Función para extraer el número de la cadena
def extraer_id_player(cadena):
    match = re.search(r'/(\d+)$', cadena)
    if match:
        return int(match.group(1))
    else:
        return None

# Aplicar la función a cada fila de la columna
df_players['player_id'] = df_players['href'].apply(extraer_id_player)
df_players.drop('href',axis=1,inplace=True)
df_players.head()
```

```python
# Función para extraer el número de la clave 'href'
def extraer_id_club(diccionario):
    # Utilizar una expresión regular para buscar el patrón '/verein/<número>/'
    match = re.search(r'/verein/(\d+)/', diccionario['seasoned_href'])
    if match:
        return int(match.group(1))
    else:
        return None


# Aplicar la función a cada fila de la columna 'diccionarios'
df_players['club_id'] = df_players['parent'].apply(extraer_id_club)
df_players.drop('parent',axis=1,inplace=True)
df_players.head()
```

Eliminando columnas con la totalidad de los datos nulos

```python
df_players.drop('day_of_last_contract_extension',axis=1,inplace=True)
df_players.drop('market_value_history',axis=1,inplace=True)
df_players.drop('current_market_value',axis=1,inplace=True)
df_players.drop('highest_market_value',axis=1,inplace=True)
df_players.head()
```

Eliminando columnas con datos redundantes o con direcciones web

```python
df_players.drop('type',axis=1,inplace=True)
df_players.drop('player_agent',axis=1,inplace=True)
df_players.drop('image_url',axis=1,inplace=True)
df_players.drop('current_club',axis=1,inplace=True)
df_players.drop('social_media',axis=1,inplace=True)

df_players.head()
```

Eliminand columnas con datos de nula importancia

```python
df_players.drop('name_in_home_country',axis=1,inplace=True)
df_players.head()
```

```python
#Elimina la palabra 'Seats', los "." y se transofrma a entero
df_players['number'] = df_players['number'].fillna('0').str.replace('#', '').astype(int)
df_players['height'] = df_players['height'].str.replace(' m', '').str.replace('\xa0m', '').str.replace(',', '.').astype(float)
df_players['joined'] = df_players['joined'].str.strip().str.replace('-', '')
df_players['contract_expires'] = df_players['contract_expires'].str.strip().str.replace('-', '')

# Convertir la columna a formato datetime
df_players['date_of_birth'] = pd.to_datetime(df_players['date_of_birth'], format='%b %d, %Y')
df_players['joined'] = pd.to_datetime(df_players['joined'], format='%b %d, %Y')
df_players['contract_expires'] = pd.to_datetime(df_players['contract_expires'], format='%b %d, %Y')
df_players.head()
```

```python
# Usar pd.json_normalize() para transformar los diccionarios en columnas
df_info = pd.json_normalize(df_players['place_of_birth'])

# Renombrar las columnas según tus necesidades
df_info.rename(columns={'country': 'country_of_birth', 'city': 'city_of_birth'}, inplace=True)

# Eliminar espacios adicionales en los valores de la columna 'city_of_birth'
df_info['city_of_birth'] = df_info['city_of_birth'].str.strip()

# Concatenar las nuevas columnas al DataFrame original
df_players = df_players.drop(columns=['place_of_birth']).join(df_info)

df_players.head()
```

```python
df_players.to_csv('./Data/liga brasil/players.csv',index=False)
```
