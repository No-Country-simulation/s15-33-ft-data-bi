```python
# Crear un DataFrame vacío para almacenar todos los eventos
df_home_lineups = pd.DataFrame()

# Iterar sobre las listas de eventos y otra columna en df_games
for index, row in df_game_lineups.iterrows():
    list_events = row['events']
    game_id = row['game_id']
    
    # Convertir la lista de eventos en un DataFrame
    df = pd.DataFrame(list_events)
    
    # Agregar la otra columna a df
    df['game_id'] = game_id
    
    # Concatenar el DataFrame de eventos actual con el DataFrame general de eventos
    df_home_lineups = pd.concat([df_home_lineups, df], ignore_index=True)

df_home_lineups.head()
```

```python
# Usar pd.json_normalize() para transformar los diccionarios en columnas
df_info = pd.json_normalize(df_game_lineups.drop('parent',axis=1,inplace=True)['home_club'])

df_info.head()
```
